## Seaborn



### Distribution Plots

- distplot
- joinplot
- pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

### Correlation with heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr(),)
```

### Bivariate analysis

```python
sns.jointplot(x=df['tip'],y=df['total_bill'],kind='hex')
```

```python
sns.jointplot(x=df['tip'],y=df['total_bill'],kind='reg')
```

### Pairplot

kind of a scatter plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

### Displot

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

## Categorical Plot

Seaborn also helps us in doing analysis on Categorical data points.

- boxplot
- vilionplot
- countplot
- barplot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
##bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

### Boxplot

```python
sns.boxplot('smoker','tip',data=df)
```

```python
sns.boxplot(data=df, orient='h')
```

```python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
## Vilionplot


sns.violinplot(x='total_bill',y='day',hue='sex',color='r',data=df)
```
