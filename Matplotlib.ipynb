### Matplotlib

Matplotlib is a plotting library for the Python programming language and it is numerical mathematical extension Numpy.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,30)
y=np.arange(20,50)
```

```python
 #scatter plot
    
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D graph')
plt.savefig('2dtest.jpg')
```

```python
#plt plot

y=x**2
plt.plot(x,y,'r--',linewidth=2,markersize=10)
plt.xlabel('x-azis')
plt.ylabel('y-axis')
```

```python
# creating subplots

plt.subplot(2,2,1)   #(no of rows,no of col,postions)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
```

```python
## barplot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,5,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.xlabel('x_axis')
```

```python
## Histograms

a=np.array([22,87,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=12)
plt.title("histogram")
```

```python
## Boxplot

data=[np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

```python
#pie chart

label='Python','c++','java','ruby'
size=[316,288,349,100]
colors=['yellow','red','green','gold']
explod=(0.2,0,0,0)

plt.pie(size,explode=explod,labels=label,colors=colors,
autopct='%1.1f%%',shadow=False)
plt.axis('equal')
```
